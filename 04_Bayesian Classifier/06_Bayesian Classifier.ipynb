Iris Dataset consists of 150 samples of iris flowers, each with four features: sepal length, sepal width, petal length, and petal width. The goal is to classify each iris sample into one of three species: setosa, versicolor, or virginica. Using the Bayesian classifier algorithm with the Iris dataset, a model is created to predict the species of iris flowers based on their features. 

To enhance the model performance of Naive Bayes model for predicting wine categories, feature scaling was applied using StandardScaler(). The model achieved an accuracy score of 0.9722.

Checked thed is there any chance for overfitting by comparing train_accuracy and test_accuracy 

### Step 1: Import Required Libraries

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report 
from sklearn.feature_selection import SelectFromModel
```

### Step 2: Load Dataset:

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
y
```

### Step 3: Split Dataset:

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Step 4: Feature scaling

```python
# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Step 5: Create GaussianNB Classifier:

```python
# Initialize GaussianNB Classifier
naive_bayes = GaussianNB()
```

### Step 6: Train the Model:

```python
# Train the model
naive_bayes.fit(X_train, y_train)
```

### Step 7: Make Predictions:

```python
# Make predictions
y_pred = naive_bayes.predict(X_test)
```

### Step 8: Evaluate the Model:

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

Class A: All 19 instances of Class A were correctly classified (TP).
Class B: 12 instances of Class B were correctly classified (TP), 1 instance of Class B was incorrectly classified as Class C (FP), resulting in a false negative (FN) for Class B.
Class C: All 13 instances of Class C were correctly classified (TP).

```python
# Generate classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

Checking model is not overfitted or not

```python

# Make predictions on training data
y_train_pred = naive_bayes.predict(X_train_scaled)

# Make predictions on testing data
y_test_pred = naive_bayes.predict(X_test_scaled)

# Calculate accuracy on training data
train_accuracy = accuracy_score(y_train, y_train_pred)

# Calculate accuracy on testing data
test_accuracy = accuracy_score(y_test, y_test_pred)

# Check if the model is overfitting
if train_accuracy > test_accuracy:
    print("The model is overfitting.")
else:
    print("The model is not overfitting.")
```
